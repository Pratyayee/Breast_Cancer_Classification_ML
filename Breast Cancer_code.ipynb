```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

Xs = scale(X)

Xs_train, Xs_test, y_train, y_test = train_test_split(Xs, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

    
knn = KNeighborsClassifier(n_neighbors = 7)
knn_model = knn.fit(Xs_train, y_train)

    
```

```python
print('Training Acuuracy for knn: %f' % knn_model.score(Xs_train, y_train))
```

```python
y_pred = knn.predict(Xs_test)


from sklearn import metrics


print("Testing Accuracy for knn:",metrics.accuracy_score(y_test, y_pred))
```

```python

from sklearn import svm


svc = svm.SVC(kernel='linear') # Linear Kernel


svc.fit(Xs_train, y_train)

y_pred = svc.predict(Xs_test)
```

```python
print('Training Accuracy for SVM(linear): %f' % svc.score(Xs_train, y_train))
```

```python

from sklearn import metrics


print("Testing Accuracy for SVM(Linear):",metrics.accuracy_score(y_test, y_pred))
```

```python

from sklearn import svm


svc1 = svm.SVC(kernel='rbf') # rbf Kernel

svc1.fit(Xs_train, y_train)


y_pred1 = svc1.predict(Xs_test)
```

```python
print('Training Accuracy for SVM(rbf): %f' % svc1.score(Xs_train, y_train))
```

```python

from sklearn import metrics

print("Testing Accuracy for SVM(rbf):",metrics.accuracy_score(y_test, y_pred1))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(random_state=0)

mlp.fit(Xs_train, y_train)

y_pred2 = mlp.predict (Xs_test)
```

```python
print('Training Accuracy for MLP: %f' % mlp.score(Xs_train, y_train))
```

```python

from sklearn import metrics


print("Testing Accuracy for mlp:",metrics.accuracy_score(y_test, y_pred2))
```

```python
print(" Here, from the above result we get that The Accuracy of MLP is maximum according to Testing Score.  ")
```
